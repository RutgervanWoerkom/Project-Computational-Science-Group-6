```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
import cv2
```

```python
"""
0: Ocean
1: Forest
2: Scrubland
3: Woodland
4: Grassland
5: Desert
"""
x_scale = 1100
y_scale = 900

im = cv2.imread("../../datasets/raw/corrected_vegetation_australia2.png", 0)
im = np.flip(im, 0)

# add extra ocean buffer
im = np.concatenate((np.zeros((30, 500)), im, np.zeros((15, 500))), 0)
im = cv2.resize(im, (500, 500), interpolation=cv2.INTER_NEAREST)

im = np.concatenate((np.zeros((500, 20)), im, np.zeros((500, 20))), 1)
im = cv2.resize(im, (x_scale, y_scale), interpolation=cv2.INTER_NEAREST)

# Assign colors to labels
im = np.where(im != 255, im, 0)
im = np.where(im != 101, im, 1)
im = np.where(im != 122, im, 2)
im = np.where(im != 209, im, 3)
im = np.where(im != 226, im, 4)
im = np.where(im != 232, im, 5)

assert len(np.unique(im)) == 6

figure(num=None, figsize=(18.88, 16))
plt.imshow(im, cmap='hot', interpolation='nearest', origin='lower')
np.save('../../datasets/processed/australia-vegetation', im)
plt.show()
```

```python
x_scale = 1100
y_scale = 900

im = cv2.imread("../../datasets/raw/map-australia.png", 0)
im = np.flip(im, 0)

shape = np.shape(im)
print(shape)



# Assign colors to labels
print(np.unique(im))
im = cv2.inRange(im, 208, 220)

im = im - 255
# im = im * (1/255)
im = im.astype(int)

# add extra ocean buffer
# bottom top
im = np.concatenate((np.zeros((85, shape[1])), im, np.zeros((9, shape[1]))), 0)
im = cv2.resize(im, (2000, 2000), interpolation=cv2.INTER_NEAREST)

# left right
im = np.concatenate((np.zeros((2000, 55)), im, np.zeros((2000, 85))), 1)
im = cv2.resize(im, (x_scale, y_scale), interpolation=cv2.INTER_NEAREST)

print(np.unique(im))

figure(num=None, figsize=(18.88, 16))
plt.imshow(im, cmap='hot', interpolation='nearest', origin='lower')
np.save('../../datasets/processed/australia-mask', im)
plt.show()
# print(np.shape(im))
```
